### Exercício 8
Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
df.head()
```

```python
print(df['Pressão sanguínea'].dtype)
```

```python
def separar_pressao(pressao):
    sistolica, diastolica = pressao.split('/')
    return pd.Series([int(sistolica), int(diastolica)], index=['Sistolica', 'Diastolica'])
```

```python
df[['Sistolica', 'Diastolica']] = df['Pressão sanguínea'].apply(separar_pressao)
df.head()
```

```python
media_por_genero = df.groupby('Gênero').agg({'Sistolica': 'mean', 'Diastolica': 'mean'})
media_por_genero
```

# Resposta
### Após realizar a média dos valores presentes na coluna de pressão sanguínea, podemos visualizar que as Mulheres têm maior pressão sanguínea média.

# Exercício 10
Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
media_passos_acima_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_passos_abaixo_70 = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()
media_passos_acima_70 > media_passos_abaixo_70
```
